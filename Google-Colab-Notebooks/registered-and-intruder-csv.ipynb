```python
import os
import shutil
import pandas as pd

# Define paths
csv_path = '/content/drive/My Drive/mindlock-eeg-files/session-1/registered-and-intruder.csv'
base_directory = '/content/drive/My Drive/mindlock-eeg-files/session-1/preprocessed-sub/'
output_directory = '/content/drive/My Drive/mindlock-eeg-files/session-1/demo-testing/'

# Create output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Load CSV
df = pd.read_csv(csv_path)

# Mapping for task codes
task_mapping = {3: 'VEP', 4: 'AEP', 5: 'SEP'}

# Iterate over rows in the CSV
for _, row in df.iterrows():
    file_name = row['file_name']
    subject_number = row['subject_number']
    task_code = row['task_code']

    # Remove "-epo" from file_name to get the correct name
    correct_file_name = file_name.replace('-epo', '')

    # Get task folder
    task_folder = task_mapping.get(task_code, None)
    if task_folder is None:
        print(f"Invalid task code {task_code} for file {file_name}. Skipping...")
        continue

    # Get subject folder
    subject_folder = f"sub{str(subject_number).zfill(3)}"

    # Get the inner folder format
    inner_folder = f"{subject_folder}-{task_folder}-unfiltered"

    # Construct the full path to the correct file
    file_path = os.path.join(base_directory, task_folder, subject_folder, inner_folder, correct_file_name)

    # Check if the file exists
    if os.path.exists(file_path):
        # Copy the file to the output directory
        destination_path = os.path.join(output_directory, correct_file_name)
        shutil.copy(file_path, destination_path)
        print(f"Copied {file_path} to {destination_path}.")
    else:
        print(f"File not found: {file_path}. Skipping...")

print("Task completed!")
```

```python
import os
import pandas as pd

# Define paths
fif_directory = '/content/drive/My Drive/mindlock-eeg-files/session-2/for-testing/manual-testing/'
csv_path = '/content/drive/My Drive/eeg-biometric-competition/Calibration_Info.csv'
output_csv_path = '/content/drive/My Drive/mindlock-eeg-files/session-2/for-testing/manual-testing.csv'

# Load the Calibration_Info.csv
calibration_info = pd.read_csv(csv_path)

# Initialize an empty DataFrame to store the results
results = pd.DataFrame(columns=['EpochID', 'SubjectID', 'Task', 'claimed_user'])

# Loop over all files in the directory
for filename in os.listdir(fif_directory):
    if filename.endswith('.fif'):
        # Extract the epoch name
        epoch_name = filename.replace('.fif', '')

        # Search for epoch_name in the Calibration_Info.csv
        row = calibration_info[calibration_info['EpochID'] == epoch_name]

        # If the epoch_name is found, extract required columns and append to results
        if not row.empty:
            row['claimed_user'] = row['SubjectID']

            # Select required columns and append to results
            selected_data = row[['EpochID', 'SubjectID', 'Task', 'claimed_user']]
            results = pd.concat([results, selected_data])

# Save the results to a new CSV
results.to_csv(output_csv_path, index=False)

print(f"Results saved to {output_csv_path}")
```

```python
import pandas as pd

# Path to the CSV file
output_csv_path = '/content/drive/My Drive/mindlock-eeg-files/session-2/for-testing/manual-testing.csv'

# Load the CSV
df = pd.read_csv(output_csv_path)

# Function to extract and convert the numerical part from the subject IDs
def extract_number(subject_id):
    return int(subject_id.replace("sub", ""))

# Apply the function to 'SubjectID' and 'claimed_user' columns
df['SubjectID'] = df['SubjectID'].apply(extract_number)
df['claimed_user'] = df['claimed_user'].apply(extract_number)

# Save the modified DataFrame back to the same file
df.to_csv(output_csv_path, index=False)

print(f"Updated CSV saved to {output_csv_path}")
```

```python
import random
import pandas as pd
from pathlib import Path

# Define file paths
csv_path = '/content/drive/My Drive/mindlock-eeg-files/session-2/for-testing/manual-testing.csv'
intruder_directory = '/content/drive/My Drive/mindlock-eeg-files/session-2/for-testing/intruder'
output_csv_path = '/content/drive/My Drive/mindlock-eeg-files/session-2/for-testing/registered-and-intruder.csv'

# Load the existing CSV file
df = pd.read_csv(csv_path)

# Get all intruder EEG files
intruder_files = list(Path(intruder_directory).glob('*'))
random.shuffle(intruder_files)  # Shuffle to randomly select the first 400 files

# Prepare list to collect intruder data rows
intruder_rows = []

# Select 400 files, skipping files that start with "jittered_"
for file in intruder_files:
    if file.name.startswith("jittered_"):
        continue  # Skip files that start with "jittered_"
    intruder_rows.append({
        'EpochID': file.name,
        'SubjectID': 86,  # Intruder's subject number
        'Task': 3,        # Fixed task code for intruder data
        'claimed_user': random.randint(1, 20)  # Random claimed user from 1 to 20
    })

# Convert intruder rows to DataFrame and append to the original data
intruder_df = pd.DataFrame(intruder_rows)
df = pd.concat([df, intruder_df], ignore_index=True)

# Save the modified DataFrame to the output CSV file
df.to_csv(output_csv_path, index=False)

print(f"Modified CSV saved to {output_csv_path}")
```

```python
import shutil
from pathlib import Path

# Define directories
intruder_directory = '/content/drive/My Drive/mindlock-eeg-files/session-2/for-testing/intruder'
source_directory = '/content/drive/My Drive/mindlock-eeg-files/session-2/filtered-epoch-sub'
task_codes = ['VEP', 'AEP', 'SEP']
subject_nums = [91, 92, 93]

# Initialize counter for copied files
file_count = 0

# Ensure the intruder directory exists
Path(intruder_directory).mkdir(parents=True, exist_ok=True)

# Copy files
for task in task_codes:
    task_dir = Path(source_directory) / task
    for subject_num in subject_nums:
        subject_folder = task_dir / f"sub{str(subject_num).zfill(3)}"
        if subject_folder.exists():
            for file in subject_folder.iterdir():
                if file.is_file():  # Check if it's a file
                    shutil.copy(file, intruder_directory)
                    file_count += 1

print(f"Total files copied: {file_count}")
```
